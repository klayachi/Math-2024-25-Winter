## **25. Differential Equations**
1. Solve the following first-order ordinary differential equations:
   - $y'(x)= y$
   - $y'(x) = \frac{1}{2y(x)}$
  
2. Solve the first-order ordinary differential equations using the method of separation of variables for $x$ and $y=y(x)$:

   - $\frac{dy}{dx} = \frac{x}{y}$

   - $\frac{dy}{dx} = xy$

   * $y''(x) + y'(x) = 0$, with boundary conditions $y(0) = 2$ and $y'(0) = -1$

$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

### **1. Solve the following first-order ordinary differential equations**

#### **(a) $y'(x) = y$**

The differential equation is:  
$
\frac{dy}{dx} = y
$

Separating variables:  
$
\frac{1}{y} \, dy = dx
$

Integrating both sides:  
$
\ln|y| = x + C
$

Exponentiating both sides:  
$
y = Ce^x
$

where $C$ is the constant of integration.

---

#### **(b) $y'(x) = \frac{1}{2y(x)}$**

The differential equation is:  
$
\frac{dy}{dx} = \frac{1}{2y}
$

Separating variables:  
$
2y \, dy = dx
$

Integrating both sides:  
$
y^2 = \frac{x}{2} + C
$

Taking the square root:  
$
y = \pm\sqrt{\frac{x}{2} + C}
$

where $C$ is the constant of integration.

---

### **2. Solve the first-order ordinary differential equations using the method of separation of variables**

#### **(a) $\frac{dy}{dx} = \frac{x}{y}$**

The differential equation is:  
$
\frac{dy}{dx} = \frac{x}{y}
$

Rearranging terms:  
$
y \, dy = x \, dx
$

Integrating both sides:  
$
\frac{y^2}{2} = \frac{x^2}{2} + C
$

Simplify:  
$
y^2 = x^2 + 2C
$

Taking the square root:  
$
y = \pm\sqrt{x^2 + 2C}
$

where $C$ is the constant of integration.

---

#### **(b) $\frac{dy}{dx} = \frac{y}{x}$**

The differential equation is:  
$
\frac{dy}{dx} = \frac{y}{x}
$

Separating variables:  
$
\frac{1}{y} \, dy = \frac{1}{x} \, dx
$

Integrating both sides:  
$
\ln|y| = \ln|x| + C
$

Simplify:  
$
y = Cx
$

where $C$ is the constant of integration.

---

#### **(c) $\frac{dy}{dx} = xy$**

The differential equation is:  
$
\frac{dy}{dx} = xy
$

Rearranging terms:  
$
\frac{1}{y} \, dy = x \, dx
$

Integrating both sides:  
$
\ln|y| = \frac{x^2}{2} + C
$

Exponentiating both sides:  
$
y = e^C e^{x^2 / 2}
$

Simplify:  
$
y = Ce^{x^2 / 2}
$

where $C = e^C$ is the constant of integration.

### **3. Solve the second-order ordinary differential equations**

#### **(a) $y''(x) + y'(x) = 0$, with $y(0) = 2$ and $y'(0) = -1$**

The differential equation is:  
$
y'' + y' = 0
$

Let $y'(x) = v(x)$, so $y''(x) = v'(x)$.  
The equation becomes:  
$
v'(x) + v(x) = 0
$

Separating variables:  
$
\frac{1}{v} \, dv = -dx
$

Integrating:  
$
\ln|v| = -x + C_1
$

Exponentiating:  
$
v = C_2 e^{-x}
$

Since $v = y'$, we have:  
$
y'(x) = C_2 e^{-x}
$

Integrating again:  
$
y(x) = -C_2 e^{-x} + C_3
$

Using $y(0) = 2$ and $y'(0) = -1$, solve for $C_2$ and $C_3$:  
$
y(0) = -C_2 + C_3 = 2 \quad \text{and} \quad y'(0) = C_2 = -1
$

Thus, $C_2 = -1$ and $C_3 = 1$. The solution is:  
$
y(x) = e^{-x} + 1
$

---

#### **(b) $y''(x) - y(x) = 0$, with $y(0) = 2$ and $y'(0) = 0$**

The differential equation is:  
$
y'' - y = 0
$

The characteristic equation is:  
$
r^2 - 1 = 0
$

Roots are:  
$
r = \pm 1
$

The general solution is:  
$
y(x) = C_1 e^x + C_2 e^{-x}
$

Using $y(0) = 2$ and $y'(0) = 0$, solve for $C_1$ and $C_2$:  
$
y(0) = C_1 + C_2 = 2
$

$
y'(x) = C_1 e^x - C_2 e^{-x}, \quad y'(0) = C_1 - C_2 = 0
$

From $C_1 - C_2 = 0$, we get $C_1 = C_2 = 1$. The solution is:  
$
y(x) = e^x + e^{-x}
$

---

#### **(c) $\frac{d^2 y(x)}{dx^2} = -\omega^2 y(x)$**

The differential equation is:  
$
y'' = -\omega^2 y
$

The characteristic equation is:  
$
r^2 + \omega^2 = 0
$

Roots are:  
$
r = \pm i\omega
$

The general solution is:  
$
y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x)
$

### **4. Check if $\psi(t, x) = A \cos(\omega t + kx)$ is a solution of the wave equation**

The wave equation is:  
$
\frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0
$

Compute second derivatives of $\psi$:  
$
\frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx)
$

$
\frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx)
$

Substitute into the wave equation:  
$
-A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx)) = 0
$

Factor out $\cos(\omega t + kx)$:  
$
-A (\omega^2 - v^2 k^2) = 0
$

Since $A \neq 0$, the condition $\omega^2 = v^2 k^2$ holds.  
Thus, $\psi(t, x)$ satisfies the wave equation.
