## 1. Compute

### (a) $\int 1 \, dx$

we will use $\int x^n \, dx =  \frac{1}{n+1} x^{n+1} + C$

$$
\int 1 \, dx = \int x^0 \, dx = \frac{1}{0+1} x^{0+1} + C = x + C
$$


---

### (b) $\int (x^2 + 2) \, dx$
$$
\int (x^2 + 2) \, dx = \int x^2 \, dx + \int 2 \, dx
$$
$$
= \frac{x^3}{3} + 2x + C
$$

---

### (c) $\int 2 \sin(x) \, dx$
$$
\int 2 \sin(x) \, dx = 2 \int \sin(x) \, dx
$$
$$
= -2 \cos(x) + C
$$

---

### (d) $\int \frac{3}{x} \, dx$
$$
\int \frac{3}{x} \, dx = 3 \int \frac{1}{x} \, dx
$$
$$
= 3 \ln|x| + C
$$

---

### (e) $\int \frac{1}{x^2} \, dx$
$$
\int \frac{1}{x^2} \, dx = \int x^{-2} \, dx
$$
$$
= -\frac{1}{x} + C
$$

---

### (f) $\int \left( \frac{1}{3}x^4 - 5 \right) \, dx$
$$
\int \left( \frac{1}{3}x^4 - 5 \right) \, dx = \frac{1}{3} \int x^4 \, dx - \int 5 \, dx
$$
$$
= \frac{1}{15}x^5 - 5x + C
$$

---

### (g) $\int (\sin^2 x + \cos^2 x) \, dx$
Using the trigonometric identity $\sin^2(x) + \cos^2(x) = 1$:
$$
\int (\sin^2 x + \cos^2 x) \, dx = \int 1 \, dx
$$
$$
= x + C
$$

---

### (h) $\int (5 \sin x + 3e^x) \, dx$
$$
\int (5 \sin x + 3e^x) \, dx = 5 \int \sin x \, dx + 3 \int e^x \, dx
$$
$$
= -5 \cos x + 3e^x + C
$$

---

### (i) $\int \sqrt[3]{x} \, dx$
$$
\int \sqrt[3]{x} \, dx = \int x^{1/3} \, dx
$$
$$
= \frac{3}{4}x^{4/3} + C
$$

---

### (j) $\int \sqrt{10x} \, dx$
$$
\int \sqrt{10x} \, dx = \int \sqrt{10} \cdot \sqrt{x} \, dx = \sqrt{10} \int x^{1/2} \, dx
$$
$$
= \sqrt{10} \cdot \frac{2}{3}x^{3/2} + C
$$

---

### (k) $\int \cos\left(\frac{5}{2}x + 3\right) \, dx$
Using substitution:  
Let $u = \frac{5}{2}x + 3$, then $\frac{du}{dx} = \frac{5}{2} \implies dx = \frac{2}{5}du$
$$
\int \cos\left(\frac{5}{2}x + 3\right) \, dx = \frac{2}{5} \int \cos(u) \, du
$$
$$
= \frac{2}{5} \sin(u) + C
$$
Substitute back $u$:
$$
= \frac{2}{5} \sin\left(\frac{5}{2}x + 3\right) + C
$$

---

### (l) $\int \frac{\cos(\ln(x))}{x} \, dx$
Using substitution:  
Let $u = \ln(x)$, then $\frac{du}{dx} = \frac{1}{x} \implies dx = x \, du$
$$
\int \frac{\cos(\ln(x))}{x} \, dx = \int \cos(u) \, du
$$
$$
= \sin(u) + C
$$
Substitute back $u$:
$$
= \sin(\ln(x)) + C
$$

---

### (m) $\int x \ln(x) \, dx$
Using integration by parts:  
Let $u = \ln(x)$ and $dv = x \, dx$. Then $du = \frac{1}{x}dx$ and $v = \frac{x^2}{2}$
$$
\int x \ln(x) \, dx = uv - \int v \, du
$$
$$
= \frac{x^2}{2} \ln(x) - \int \frac{x^2}{2} \cdot \frac{1}{x} \, dx
$$
$$
= \frac{x^2}{2} \ln(x) - \frac{1}{2} \int x \, dx
$$
$$
= \frac{x^2}{2} \ln(x) - \frac{x^2}{4} + C
$$

---

### (n) $\int x e^x \, dx$
Using integration by parts:  
Let $u = x$ and $dv = e^x \, dx$. Then $du = dx$ and $v = e^x$
$$
\int x e^x \, dx = uv - \int v \, du
$$
$$
= x e^x - \int e^x \, dx
$$
$$
= x e^x - e^x + C
$$
$$
= e^x(x - 1) + C
$$

## 2. Calculate integrals over the interval $[0, \pi]$ and visualize them in GeoGebra.

### (a) $f(x) = 2x + 1$
The definite integral is:
$$
\int_0^\pi (2x + 1) \, dx = \left[ x^2 + x \right]_0^\pi
$$
Evaluate at the bounds:
$$
= (\pi^2 + \pi) - (0^2 + 0)
$$
$$
= \pi^2 + \pi
$$

---

### (b) $g(x) = x^2$
The definite integral is:
$$
\int_0^\pi x^2 \, dx = \left[ \frac{x^3}{3} \right]_0^\pi
$$
Evaluate at the bounds:
$$
= \frac{\pi^3}{3} - \frac{0^3}{3}
$$
$$
= \frac{\pi^3}{3}
$$

## 3. Calculate the area of the region bounded by the lines:
$x = 1$, $x = 2$, $y = 0$, and $y = x^2 + 1$.

The area is given by the definite integral:
$$
\text{Area} = \int_1^2 (x^2 + 1) \, dx
$$
Solve the integral:
$$
\int (x^2 + 1) \, dx = \frac{x^3}{3} + x
$$
Evaluate at the bounds:
$$
\text{Area} = \left[ \frac{x^3}{3} + x \right]_1^2
$$
$$
= \left( \frac{2^3}{3} + 2 \right) - \left( \frac{1^3}{3} + 1 \right)
$$
$$
= \left( \frac{8}{3} + 2 \right) - \left( \frac{1}{3} + 1 \right)
$$
$$
= \frac{8}{3} + \frac{6}{3} - \frac{1}{3} - \frac{3}{3}
$$
$$
= \frac{10}{3}
$$

## 4. Calculate the area under the sine curve over $[0, \pi]$:
The area is given by:
$$
P = \int_0^\pi \sin(x) \, dx
$$
Solve the integral:
$$
\int \sin(x) \, dx = -\cos(x)
$$
Evaluate at the bounds:
$$
P = \left[ -\cos(x) \right]_0^\pi
$$
$$
= -\cos(\pi) + \cos(0)
$$
$$
= -(-1) + 1
$$
$$
= 2
$$

## 6. Find the distance of a moving particle for $x(t) = 3t^2 - 6t + 1$ over $t \in [0, 2]$:
The distance is given by:
$$
\text{Distance} = \int_0^2 |v(t)| \, dt
$$
where $v(t) = \frac{dx}{dt} = 6t - 6$.

### Step 1: Find critical points of $v(t)$
Solve $v(t) = 0$:
$$
6t - 6 = 0 \implies t = 1
$$

### Step 2: Split the integral at $t = 1$
For $t \in [0, 1]$, $v(t) < 0$, so $|v(t)| = -(v(t)) = -6t + 6$.  
For $t \in [1, 2]$, $v(t) > 0$, so $|v(t)| = v(t) = 6t - 6$.

### Step 3: Compute the integral
$$
\text{Distance} = \int_0^1 (-6t + 6) \, dt + \int_1^2 (6t - 6) \, dt
$$
First integral:
$$
\int_0^1 (-6t + 6) \, dt = \left[ -3t^2 + 6t \right]_0^1
$$
$$
= (-3(1)^2 + 6(1)) - (-3(0)^2 + 6(0))
$$
$$
= -3 + 6 = 3
$$
Second integral:
$$
\int_1^2 (6t - 6) \, dt = \left[ 3t^2 - 6t \right]_1^2
$$
$$
= (3(2)^2 - 6(2)) - (3(1)^2 - 6(1))
$$
$$
= (12 - 12) - (3 - 6)
$$
$$
= 0 + 3 = 3
$$

### Step 4: Total distance
$$
\text{Distance} = 3 + 3 = 6
$$
