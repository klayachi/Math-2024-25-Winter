   $
   \int 1 \, dx = x + C

2. **Integral of $ x^2 + 2 $**:

   $
   \int (x^2 + 2) \, dx = \frac{x^3}{3} + 2x + C
   $

3. **Integral of $ 2\sin(x) $**:

   $
   \int 2\sin(x) \, dx = -2\cos(x) + C
   $

4. **Integral of $ \frac{3}{x} $**:

   $
   \int \frac{3}{x} \, dx = 3 \ln|x| + C
   $

5. **Integral of $ \frac{1}{x^2} $**:

   $
   \int \frac{1}{x^2} \, dx = -\frac{1}{x} + C
   $

6. **Integral of $ \frac{1}{3}x^4 - 5 $**:

   $
   \int \left( \frac{1}{3}x^4 - 5 \right) \, dx = \frac{x^5}{15} - 5x + C
   $

7. **Integral of $ \sin^2(x) + \cos^2(x) $**:

   Using the identity $ \sin^2(x) + \cos^2(x) = 1 $:
  
   $
   \int (\sin^2(x) + \cos^2(x)) \, dx = \int 1 \, dx = x + C
   $

8. **Integral of $ 5 \sin(x) + 3e^x $**:

   $
   \int (5 \sin(x) + 3e^x) \, dx = -5\cos(x) + 3e^x + C
   $

9. **Integral of $ \sqrt[3]{x} $**:

   $
   \int \sqrt[3]{x} \, dx = \int x^{\frac{1}{3}} \, dx = \frac{3}{4}x^{\frac{4}{3}} + C
   $

10. **Integral of $ \sqrt{10x} $**:

    $\int \sqrt{10x} \, dx = \int \sqrt{10}x^{\frac{1}{2}} \, dx = \frac{2\sqrt{10}}{3}x^{\frac{3}{2}} + C$

11. **Integral of $ \cos\left(\frac{5}{2}x + 3\right) $**:
    
    $
    \int \cos\left(\frac{5}{2}x + 3\right) \, dx = \frac{2}{5} \sin\left(\frac{5}{2}x + 3\right) + C
    $

12. **Integral of $ \frac{\cos(\ln(x))}{x} $**:

    This is a standard integral:
    
    $
    \int \frac{\cos(\ln(x))}{x} \, dx = \sin(\ln(x)) + C
    $

13. **Integral of $ x \ln(x) $**:

    We use integration by parts:
    
    $
    \int x \ln(x) \, dx = \frac{x^2}{2} \ln(x) - \frac{x^2}{4} + C
    $

14. **Integral of $ x e^x $**:

    Using integration by parts:
    
    $
    \int x e^x \, dx = (x - 1)e^x + C
    $

$= \frac{\pi^3}{3} - \frac{0^3}{3} = \frac{\pi^3}{3}$

## **Exercise 3**
The area is bounded vertically by $y = 0$ (the x-axis) and $y = x^2 + 1$, and horizontally by $x = 1$ and $x = 2$. The area can be found by integrating the function $y = x^2 + 1$ from $x = 1$ to $x = 2$.

The area $A$ is given by:

$A = \int_{1}^{2} (x^2 + 1) \, dx$

$\int (x^2 + 1) \, dx = \frac{x^3}{3} + x + C$

$A = \left[ \frac{x^3}{3} + x \right]_{1}^{2}$

$= \left( \frac{2^3}{3} + 2 \right) - \left( \frac{1^3}{3} + 1 \right)$

$= \left( \frac{8}{3} + 2 \right) - \left( \frac{1}{3} + 1 \right)$

$= \left( \frac{8}{3} + \frac{6}{3} \right) - \left( \frac{1}{3} + \frac{3}{3} \right)$

$= \frac{14}{3} - \frac{4}{3}$

$= \frac{10}{3}$

## **Exercise 6**
$t=0$, $t=2$,position function: $x(t) = 3t^2 - 6t + 1$, Distance= ?
### Step 1: Compute the Velocity

The velocity:

$v(t) = \frac{d}{dt} \left( 3t^2 - 6t + 1 \right)$

Taking the derivative:

$v(t) = 6t - 6$

### Compute the Speed

The speed is the absolute value of the velocity:

$\text{Speed} = |v(t)| = |6t - 6|$

### the Points Where Velocity Changes Sign

$6t - 6 = 0$$:

$6t = 6 \quad \Rightarrow \quad t = 1$

So, the velocity changes sign at $t = 1$.

### the Distance Traveled

The total distance traveled is given by the integral of the speed over the time interval $[0, 2]$. Since the velocity changes sign at $t = 1$, we split the integral at $t = 1$:

$\text{Distance} = \int_0^1 |6t - 6| \, dt + \int_1^2 |6t - 6| \, dt$

- For $t \in [0, 1]$, $6t - 6$ is negative, so $|6t - 6| = -(6t - 6) = 6 - 6t$.
- For $t \in [1, 2]$, $6t - 6$ is positive, so $|6t - 6| = 6t - 6$.

Thus, the distance is:

$\text{Distance} = \int_0^1 (6 - 6t) \, dt + \int_1^2 (6t - 6) \, dt$

$\int_0^1 (6 - 6t) \, dt = \left[ 6t - 3t^2 \right]_0^1 = \left( 6(1) - 3(1)^2 \right) - (0) = 6 - 3 = 3$

$\int_1^2 (6t - 6) \, dt = \left[ 3t^2 - 6t \right]_1^2 = \left( 3(2)^2 - 6(2) \right) - \left( 3(1)^2 - 6(1) \right)$

$= \left( 3(4) - 12 \right) - \left( 3(1) - 6 \right) = (12 - 12) - (3 - 6) = 0 + 3 = 3$

The total distance traveled is:

$\text{Distance} = 3 + 3 = 6$
