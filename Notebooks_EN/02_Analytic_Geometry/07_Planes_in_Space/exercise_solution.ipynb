## Task 1: Plane Passing Through Points $A(1, 2, 3)$, $B(3, 4, 5)$, and $C(2, 1, 4)$

1. **Find Two Vectors**:
   $$
   \vec{AB} = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2)
   $$
   $$
   \vec{AC} = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1)
   $$

2. **Find the Normal Vector $\vec{n}$**:
   $$
   \vec{n} = \vec{AB} \times \vec{AC}
   $$
   Compute the cross product:
   $$
   \vec{n} = \begin{vmatrix}
   \hat{i} & \hat{j} & \hat{k} \\
   2 & 2 & 2 \\
   1 & -1 & 1
   \end{vmatrix} = (4 - (-2))\hat{i} - (2 - 2)\hat{j} + (2 - (-2))\hat{k}
   $$
   Simplify:
   $$
   \vec{n} = (6, 0, 4)
   $$

3. **Plane Equation**:
   The equation of the plane is:
   $$
   6(x - 1) + 0(y - 2) + 4(z - 3) = 0
   $$
   Simplify:
   $$
   6x + 4z = 18
   $$

**Final Answer**: The equation of the plane is:
$$
6x + 4z = 18
$$

## Task 4: Line of Intersection of Planes $2x + 3y + 4z = 5$ and $3x + 4y + 2z = 6$

1. **Find the Direction Vector**:
   The direction vector is the cross product of the normal vectors $\vec{n_1} = [2, 3, 4]$ and $\vec{n_2} = [3, 4, 2]$:
   $$
   \vec{d} = \vec{n_1} \times \vec{n_2} = \begin{vmatrix}
   \hat{i} & \hat{j} & \hat{k} \\
   2 & 3 & 4 \\
   3 & 4 & 2
   \end{vmatrix}
   $$
   Compute:
   $$
   \vec{d} = (3 \cdot 2 - 4 \cdot 4)\hat{i} - (2 \cdot 2 - 4 \cdot 3)\hat{j} + (2 \cdot 4 - 3 \cdot 3)\hat{k}
   $$
   Simplify:
   $$
   \vec{d} = (-10, 8, -1)
   $$

2. **Find a Point on the Line**:
   Solve the system:
   $$
   2x + 3y + 4z = 5 \quad \text{and} \quad 3x + 4y + 2z = 6
   $$
   (Substitute values to find $x, y, z$.)

**Final Answer**: The line of intersection has direction vector:
$$
\vec{d} = (-10, 8, -1)
$$

## Task 5: Plane Through $A(1, 2, 3)$ and Parallel to $\vec{v_1} = [1, 0, 1]$, $\vec{v_2} = [0, 1, -1]$

   $$

**Final Answer**: The equation of the plane is:

-x - y + z = -4

## Task 7: Distance from $A(1, 2, 3)$ to $2x + 3y + 4z = 5$

1. **Formula**:
   $$
   d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}}
   $$

2. **Simplify**:
   $$
   d = \frac{|2 + 6 + 12 - 5|}{\sqrt{4 + 9 + 16}} = \frac{15}{\sqrt{29}}
   $$

**Final Answer**:
$$
d = \frac{15}{\sqrt{29}}
$$

## Task 9: Angle Between Planes $x + y + z = 1$ and $x = 0$

1. **Normal Vectors**:
   - $\vec{n_1} = [1, 1, 1]$
   - $\vec{n_2} = [1, 0, 0]$

2. **Angle Formula**:
   $$
   \cos \theta = \frac{\vec{n_1} \cdot \vec{n_2}}{|\vec{n_1}||\vec{n_2}|}
   $$

3. **Compute**:
   $$
   \cos \theta = \frac{1}{\sqrt{3}}
   $$

**Final Answer**:
$$
\theta = \arccos\left(\frac{1}{\sqrt{3}}\right)
$$
