# Task 1: Line Through Points $A(1,2)$ and $B(3,4)$

1. **Calculate the Slope**:
   The slope of the line, $m$, is given by:
   $$
   m = \frac{y_2 - y_1}{x_2 - x_1}
   $$
   Substituting $A(1, 2)$ and $B(3, 4)$:
   $$
   m = \frac{4 - 2}{3 - 1} = \frac{2}{2} = 1
   $$

2. **Write the Equation**:
   Using the point-slope form:
   $$
   y - y_1 = m(x - x_1)
   $$
   Substitute $m = 1$, $x_1 = 1$, and $y_1 = 2$:
   $$
   y - 2 = 1(x - 1)
   $$

3. **Simplify**:
   Expand and simplify to get:
   $$
   y = x + 1
   $$

**Final Answer**: The equation of the line is $y = x + 1$.

# Task 2: Line Through $A(1,2)$ Parallel to $y = 2x + 3$

1. **Determine the Slope**:
   For parallel lines, the slopes are equal. Since the slope of $y = 2x + 3$ is $m = 2$, the required line also has $m = 2$.

2. **Write the Equation**:
   Using the point-slope form:
   $$
   y - y_1 = m(x - x_1)
   $$
   Substitute $m = 2$, $x_1 = 1$, and $y_1 = 2$:
   $$
   y - 2 = 2(x - 1)
   $$

3. **Simplify**:
   Expand and simplify:
   $$
   y = 2x
   $$

**Final Answer**: The equation of the line is $y = 2x$.

# Task 3: Line Through $A(1,2)$ Perpendicular to $y = 2x + 3$

1. **Determine the Perpendicular Slope**:
   For perpendicular lines, the slopes satisfy:
   $$
   m \cdot m_\perp = -1
   $$
   Substituting $m = 2$:
   $$
   2 \cdot m_\perp = -1 \implies m_\perp = -\frac{1}{2}
   $$

2. **Write the Equation**:
   Using the point-slope form:
   $$
   y - y_1 = m_\perp(x - x_1)
   $$
   Substitute $m_\perp = -\frac{1}{2}$, $x_1 = 1$, and $y_1 = 2$:
   $$
   y - 2 = -\frac{1}{2}(x - 1)
   $$

3. **Simplify**:
   Expand and simplify:
   $$
   y = -\frac{1}{2}x + \frac{5}{2}
   $$

**Final Answer**: The equation of the line is $y = -\frac{1}{2}x + \frac{5}{2}$.

# Task 4: Intersection of $y = 2x + 3$ and $y = 3x + 2$

1. **Find the Intersection Point**:
   Solve the system of equations:
   $$
   2x + 3 = 3x + 2
   $$
   Rearrange:
   $$
   x = 1, \, y = 5
   $$

2. **Calculate the Angle Between the Lines**:
   The angle $\theta$ between two lines with slopes $m_1$ and $m_2$ is given by:
   $$
   \tan \theta = \left| \frac{m_1 - m_2}{1 + m_1m_2} \right|
   $$
   Substituting $m_1 = 2$ and $m_2 = 3$:
   $$
   \tan \theta = \left| \frac{2 - 3}{1 + 2 \cdot 3} \right| = \frac{1}{7}
   $$

**Final Answer**:
- Intersection point: $(1, 5)$.
- Angle: $\tan^{-1}\left(\frac{1}{7}\right)$.

# Task 7: Distance From $A(1,2)$ to $y = 2x + 3$

1. **Distance Formula**:
   The distance $d$ from $(x_1, y_1)$ to a line $ax + by + c = 0$ is:
   $$
   d = \frac{|ax_1 + by_1 + c|}{\sqrt{a^2 + b^2}}
   $$
   Substituting $a = 2$, $b = -1$, $c = 3$, $x_1 = 1$, and $y_1 = 2$:
   $$
   d = \frac{|2(1) - 2 + 3|}{\sqrt{2^2 + (-1)^2}} = \frac{3}{\sqrt{5}}
   $$

**Final Answer**: The distance is $\frac{3}{\sqrt{5}}$.
